```python
# Import necessary libraries
import pandas as pd
import joblib
```

```python
# load the model, encoder and unseen data
loaded_encoder = joblib.load("Model/leave_one_out_encoder.pkl")
loaded_model = joblib.load("Model/random_forest_model.pkl")

unseen_data = pd.read_csv('E:/my files/project/food-allergen/sample.csv', keep_default_na=False, na_values = [""])
print("Data, Model and encoder loaded successfully")
```

```python
loaded_encoder
```

```python
unseen_data
```

```python
# Transform unseen data
categorical_columns = unseen_data.select_dtypes(include=['object']).columns

unseen_data_encoded = loaded_encoder.transform(unseen_data[categorical_columns])

unseen_data = pd.concat([unseen_data.drop(categorical_columns, axis=1), unseen_data_encoded], axis=1)

unseen_data
```

```python
predictions = loaded_model.predict(unseen_data)

# Add predictions to the original unseen dataset
unseen_data['Predictions'] = predictions

# Save predictions to a CSV file
output_path = 'predictions_output.csv'
unseen_data.to_csv(output_path, index=False)
print(f"Predictions saved to '{output_path}'")
```
